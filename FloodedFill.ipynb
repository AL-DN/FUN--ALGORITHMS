```python
import matplotlib.colors as mcolors

def display_grid(grid):
    # Define a custom colormap with fixed colors
    colors = ['red', 'green', 'black', 'blue']  # Example colors for values 0, 1, 2, 3
    cmap = mcolors.ListedColormap(colors)

    plt.imshow(grid, cmap=cmap, vmin=0, vmax=3)  # Use the custom colormap
    plt.show()
```

```python
# idenifies all neigboring px with same color and is in bounds
def findNeighbors(img, row, col, oColor):

  possibleNeighbors = [(row, col-1),(row,col+1),(row-1,col),(row+1,col)]
  colorNeighbors = []
  for row, col in possibleNeighbors:
    if row < 0 or col < 0 or row >= len(img) or (col >= len(img[0])):
      continue

    if img[row][col] == oColor:
     colorNeighbors.append((row,col))

  return colorNeighbors


# uses BFS in order to change colors of pxs closest to start

def change_color(G, sRow, sCol, nColor):
  img = copy.deepcopy(G)
  queue = [(sRow,sCol)]
  visited = set()
  oColor = img[sRow][sCol]

  while len(queue) > 0:
    row, col = queue.pop(0)
    visited.add((row,col))
    img[row][col] = nColor
    display_grid(img)
    for nRow, nCol in findNeighbors(img, row, col, oColor):
      if (nRow,nCol) not in visited:
        queue.append((nRow,nCol))
```

```python
change_color(G,0,1,4)
```
